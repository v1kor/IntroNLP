```python
# Import the required libraries
import requests
from bs4 import BeautifulSoup
```

```python
# ============================================
# 1. SET UP PROPER USER-AGENT (REQUIRED BY WIKIMEDIA POLICY)
# ============================================
# User-Agent identifies your bot to the web server
# Without a proper User-Agent, Wikipedia may block your requests
# Format: "BotName/Version Description"
# Include your course/project info so admins can contact you if needed
headers = {
    'User-Agent': 'EducationalBot/1.0 LearningBeautifulSoup'
}
# ============================================
# 2. USE CANONICAL URL (ONLY /wiki/ArticleName, NO QUERY PARAMETERS)
# ============================================
# Canonical URL = standard, clean URL format without extra parameters
# Good:   https://en.wikipedia.org/wiki/Natural_language_processing
# Bad:    https://en.wikipedia.org/w/index.php?title=Natural_language_processing&oldid=123456
# Why?    Canonical URLs can be served from cache (faster for you, less load on Wikipedia)
url = 'https://en.wikipedia.org/wiki/Natural_language_processing'
```

```python
# ============================================
# 3. MAKE THE HTTP REQUEST WITH HEADERS
# ============================================
# requests.get() sends an HTTP GET request to the server
# headers parameter adds your User-Agent to the request
# GZIP compression is handled automatically by requests (saves bandwidth)
r = requests.get(url, headers=headers)

# ============================================
# 4. CHECK THE RESPONSE STATUS CODE
# ============================================
# Status codes tell you what happened:
# 200 = OK - Success! Page loaded correctly
# 404 = Not Found - Page doesn't exist (check your URL)
# 403 = Forbidden - Access denied (likely missing User-Agent)
# 429 = Too Many Requests - You're requesting too fast, need to slow down
# 500 = Server Error - Wikipedia's server has a problem
print(f"Status code: {r.status_code}")  # Should be 200 for success
```

```python
r.status_code
```

```python
# ============================================
# 5. PARSE THE HTML CONTENT WITH BEAUTIFULSOUP
# ============================================
# r.text contains the HTML source code of the Wikipedia page
# BeautifulSoup parses this HTML into a structured tree we can navigate
# 'html.parser' is Python's built-in HTML parser (no extra installation needed)
soup = BeautifulSoup(r.text, 'html.parser')
```

# HOW TO ACCESS HTML TAGS IN BEAUTIFULSOUP

```python
# Find the first <h1> element in the HTML
# <h1> is a heading tag - the main title of the page (most important heading)
first_h1 = soup.find('h1')

# Check if the element exists (to avoid errors)
if first_h1:
    # Print the tag name: <h1>
    print(f"Tag name: <{first_h1.name}>")
    
    # Print the text content inside the tag
    print(f"Tag text: '{first_h1.text}'")
    
    # Print all HTML attributes (class, id, etc.)
    print(f"Tag attributes: {first_h1.attrs}")
    
    # Print the Python object type
    print(f"Tag type: {type(first_h1)}")  # <class 'bs4.element.Tag'>
```

```python
# Find ALL <h2> elements (subheadings/section titles) in the HTML
# <h2> is a secondary heading - used for main sections of the page
# .find_all() returns a LIST of all matching tags
all_h2 = soup.find_all('h2')

# Print how many h2 tags were found
print(f"Found {len(all_h2)} subheadings (h2 tags)")

# Loop through each h2 tag in the list
for i, heading in enumerate(all_h2):
    # Get the text content of the heading and strip extra whitespace
    title = heading.get_text(strip=True)
    print(f"{i+1}. {title}")
```

```python
# Get ALL tags from the entire page
all_tags = soup.find_all()

# Print how many tags total
print(f"Total tags on page: {len(all_tags)}")

# Print first 20 tag names
print("\nFirst 20 tags:")
for i, tag in enumerate(all_tags[:20]):
    print(f"{i+1}. <{tag.name}>")

# Get unique tag names only
unique_tags = set([tag.name for tag in all_tags])
print(f"\nUnique tag types: {len(unique_tags)}")
print(f"All tag types: {sorted(unique_tags)}")

# Count frequency of each tag type
from collections import Counter
tag_counts = Counter([tag.name for tag in all_tags])
print(f"\nMost common tags:")
for tag_name, count in tag_counts.most_common(10):
    print(f"  <{tag_name}>: {count} times")
```

# Extract links from Wikipedia:

```python
links_list = soup.find_all('a') # Get ALL links (all <a> tags)
print(type(links_list))
print(len(links_list))
```

```python
print(links_list[:20],sep='/n')
```

```python
# Extract just the href attribute (the URL) from each link
hrefs = [link.get('href') for link in links_list]
```

```python
# Print first 20 hrefs
print("All href attributes:")
for href in hrefs[:20]:
    print(href)
```

```python
# Get only external links (start with http)
external_links = [link.get('href') for link in links_list 
                  if link.get('href') and link.get('href').startswith('http')]

print(f"\nExternal links: {len(external_links)}")
for link in external_links[:5]:
    print(link)
```

# Find all tables on the page

```python
# Find all tables on the page
all_tables = soup.find_all('table')

print(f"Total tables found: {len(all_tables)}")

# Loop through each table
for i, table in enumerate(all_tables[:3]):  # First 3 tables
    print(f"\n=== TABLE {i+1} ===")
    
    # Get table attributes (class, id, etc.)
    print(f"Table class: {table.get('class')}")
    print(f"Table id: {table.get('id')}")
    
    # Find all rows in the table
    rows = table.find_all('tr')
    print(f"Number of rows: {len(rows)}")
    
    # Extract data from each row
    for j, row in enumerate(rows[:5]):  # First 5 rows
        # Get header cells (th) or data cells (td)
        headers = row.find_all('th')
        cells = row.find_all('td')
        
        if headers:
            row_data = [h.get_text(strip=True) for h in headers]
            print(f"  Row {j+1} (headers): {row_data}")
        elif cells:
            row_data = [c.get_text(strip=True) for c in cells]
            print(f"  Row {j+1} (data): {row_data}")
```

````python
# Импорт библиотеки NLTK для обработки естественного языка
# Import NLTK library for natural language processing
import nltk

# Скачивание модели для токенизации слов (разбивки текста на слова)
# Download the punkt tokenizer model for splitting text into words
nltk.download('punkt')

# Скачивание обновленной версии punkt для лучшей токенизации
# Download the updated punkt_tab for better tokenization
nltk.download('punkt_tab')

# Скачивание модели для определения частей речи (POS tagging)
# Download the averaged perceptron tagger for part-of-speech tagging
nltk.download('averaged_perceptron_tagger_eng')

# Скачивание списка стоп-слов (часто встречающиеся слова, которые обычно удаляют)
# Download the stopwords list (common words that are typically removed)
nltk.download('stopwords')

# Скачивание WordNet - лексической базы данных для лемматизации
# Download WordNet - a lexical database for lemmatization
nltk.download('wordnet')

# Импорт функции для разбиения текста на слова
# Import function to split text into words (tokenization)
from nltk import word_tokenize

# Импорт лемматизатора для приведения слов к нормальной форме
# Import lemmatizer to convert words to their base/dictionary form
from nltk.stem.wordnet import WordNetLemmatizer

# Импорт списка стоп-слов для их удаления из текста
# Import stopwords list to remove common words from text
from nltk.corpus import stopwords

# Импорт автокорректора для исправления орфографических ошибок
# Import autocorrect for fixing spelling mistakes
from autocorrect import Speller

# Импорт функции для разрешения смысла слов (Word Sense Disambiguation)
# Import function for word sense disambiguation (understanding word meaning in context)
from nltk.wsd import lesk

# Импорт функции для разбиения текста на предложения
# Import function to split text into sentences
from nltk.tokenize import sent_tokenize

# Импорт модуля для работы со знаками препинания
# Import module for handling punctuation characters
import string
````

````python
# Открытие и чтение файла из папки data (относительный путь)
# Open and read file from data folder (using relative path)
sentence = open("data/file.txt", "r").read()
````

````python
# Разбиение текста на отдельные слова (токенизация)
# Split text into individual words (tokenization)
words = word_tokenize(sentence)
# Вывод первых 20 слов для проверки
# Print first 20 words for verification
print(words[0:20])
````

````python
# Create an instance of the Speller class for automatic spelling correction in English
# The Speller() function initializes a spell checker that can correct misspelled words
# Parameter lang='en' specifies that we are using the English language model
# After creation, the 'spell' object can be called as a function to correct words
# Создаем экземпляр класса Speller для автоматического исправления орфографии на английском языке
# Функция Speller() инициализирует проверку орфографии, которая может исправлять слова с ошибками
# Параметр lang='en' указывает, что мы используем английскую языковую модель
# После создания объект 'spell' можно вызывать как функцию для исправления слов
spell = Speller(lang='en')
````

````python
# Call the spell checker object to correct the misspelled word 'tomtoes'
# The correct spelling should be 'tomatoes'
# The function returns the corrected version of the word
# Вызываем объект проверки орфографии для исправления слова 'tomtoes' с ошибкой
# Правильное написание должно быть 'tomatoes'
# Функция возвращает исправленную версию слова
spell('tomtoes')
````

````python
# Инициализация пустых переменных для хранения исправленного текста
# Initialize empty variables for storing corrected text
corrected_sentence = ""
corrected_word_list = []

# Цикл по каждому слову для исправления орфографических ошибок
# Loop through each word to fix spelling errors
for wd in words:
    # Проверка, является ли токен знаком препинания
    # Check if the token is a punctuation mark
    if wd in string.punctuation:
        # Добавление знака препинания без пробела перед ним
        # Add punctuation without a space before it
        corrected_sentence += wd
        # Добавление знака препинания в список слов
        # Add punctuation to the word list
        corrected_word_list.append(wd)
    else:
        # Исправление орфографии слова с помощью автокорректора
        # Correct spelling using the autocorrector
        wd_c = spell(wd)
        # Проверка, было ли слово изменено
        # Check if the word was changed
        if wd_c != wd:
            # Вывод сообщения об исправлении
            # Print correction message
            print(wd + " has been corrected to: " + wd_c)
            # Добавление исправленного слова в предложение с пробелом
            # Add corrected word to sentence with a space
            corrected_sentence += " " + wd_c
            # Добавление исправленного слова в список
            # Add corrected word to list
            corrected_word_list.append(wd_c)
        else:
            # Если слово не требует исправления, добавляем его как есть
            # If word doesn't need correction, add it as is
            corrected_sentence += " " + wd
            corrected_word_list.append(wd)

# Удаление лишнего пробела в начале и конце предложения
# Remove extra whitespace from the beginning and end of the sentence
corrected_sentence = corrected_sentence.strip()
````

## What are Stop Words? / Что такое стоп-слова?

**Stop words** are common words that are usually removed from text during preprocessing because they carry little meaningful information.

**Стоп-слова** - это распространенные слова, которые обычно удаляются из текста при предобработке, так как они несут мало значимой информации.

### Examples / Примеры:
- **English:** a, an, the, and, of, to, is, are, was, were, be, been, in, on, at, for, with, without, by, about, etc.
- **Русский:** и, в, на, с, по, к, у, о, от, из, за, под, над, перед, при, через, около, без, для, про, etc.

### Why remove stop words? / Зачем удалять стоп-слова?

| Reason | Explanation |
|--------|-------------|
| **Reduce noise** / Уменьшение шума | Stop words appear very frequently but don't help understand the main topic |
| **Save memory** / Экономия памяти | Removing them reduces dataset size by 20-35% |
| **Improve performance** / Улучшение производительности | Less data to process = faster algorithms |
| **Better features** / Лучшие признаки | Models focus on meaningful words (nouns, verbs, adjectives) |

### Example / Пример:

**Original text / Оригинальный текст:**
> "The cat is sitting on the mat and it is very comfortable."

**After stop words removal / После удаления стоп-слов:**
> "cat sitting mat comfortable"

````python
# Получение списка английских стоп-слов (a, the, and, of, to и т.д.)
# Get list of English stopwords (a, the, and, of, to, etc.)
stop_words = stopwords.words('English')
print(stop_words)
````

````python
# Создание пустого списка для слов без стоп-слов
# Create empty list for words without stopwords
corrected_word_list_without_stopwords = []

# Цикл для удаления стоп-слов из списка
# Loop to remove stopwords from the list
for wd in corrected_word_list:
    # Если слово не является стоп-словом, добавляем его в новый список
    # If word is not a stopword, add it to new list
    if wd not in stop_words:
        corrected_word_list_without_stopwords.append(wd)

# Вывод первых 20 слов после удаления стоп-слов
# Print first 20 words after stopwords removal
corrected_word_list_without_stopwords[:20]
````

````python
# Создание объекта для стемминга (Porter Stemmer) - отсекает окончания
# Create Porter Stemmer object - cuts off word endings (e.g., "running" -> "run")
stemmer = nltk.stem.PorterStemmer()

# Создание пустого списка для стеммированных слов
# Create empty list for stemmed words
corrected_word_list_without_stopwords_stemmed = []

# Цикл для применения стемминга к каждому слову
# Loop to apply stemming to each word
for wd in corrected_word_list_without_stopwords:
    # Добавление стеммированной версии слова в список
    # Add stemmed version of the word to the list
    corrected_word_list_without_stopwords_stemmed.append(stemmer.stem(wd))

# Вывод первых 20 стеммированных слов
# Print first 20 stemmed words
corrected_word_list_without_stopwords_stemmed[:20]
````

````python
# Создание объекта лемматизатора WordNet (более точный, чем стемминг)
# Create WordNet lemmatizer object (more accurate than stemming)
lemmatizer = WordNetLemmatizer()

# Создание пустого списка для лемматизированных слов
# Create empty list for lemmatized words
corrected_word_list_without_stopwords_lemmatized = []

# Цикл для применения лемматизации к каждому слову
# Loop to apply lemmatization to each word
for wd in corrected_word_list_without_stopwords:
    # Добавление лемматизированной версии слова (словарная форма)
    # Add lemmatized version of the word (dictionary form)
    corrected_word_list_without_stopwords_lemmatized.append(lemmatizer.lemmatize(wd))

# Вывод первых 20 лемматизированных слов
# Print first 20 lemmatized words
corrected_word_list_without_stopwords_lemmatized[:20]
````

````python
# Разбиение исправленного текста на отдельные предложения
# Split the corrected text into individual sentences
print(sent_tokenize(corrected_sentence))
````

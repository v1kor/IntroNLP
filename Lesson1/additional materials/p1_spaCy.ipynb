This notebook demonstrates all preprocessing steps using spaCy library.

````python
# Import the spaCy library for NLP tasks
# Импортируем библиотеку spaCy для задач обработки естественного языка
import spacy

# Import string module for punctuation handling
# Импортируем модуль string для работы со знаками препинания
import string

# Load the English language model
# Загружаем английскую языковую модель
nlp = spacy.load('en_core_web_sm')

# Print confirmation that everything is ready
# Выводим подтверждение, что всё готово
print("spaCy model loaded successfully!")
````

## Step 2: Load the text corpus

````python
# Step 2: Load the text corpus from a file
# Шаг 2: Загружаем текстовый корпус из файла

# Open and read the text file (change the path to your file)
# Открываем и читаем текстовый файл (измените путь к вашему файлу)
with open("data/file.txt", "r", encoding="utf-8") as file:
    text_corpus = file.read()

# Print first 500 characters to verify loading
# Выводим первые 500 символов для проверки загрузки
print("First 500 characters:")
print(text_corpus[:500])
````

## Step 3: Tokenization

````python
# Step 3: Apply tokenization process to the text corpus
# Шаг 3: Применяем токенизацию к текстовому корпусу

# Process the text with spaCy (creates a Doc object with tokens)
# Обрабатываем текст с помощью spaCy (создаёт объект Doc с токенами)
doc = nlp(text_corpus)

# Extract tokens (words) from the processed document
# Извлекаем токены (слова) из обработанного документа
tokens = [token.text for token in doc]

# Print first 20 tokens
# Выводим первые 20 токенов
print("First 20 tokens:")
print(tokens[:20])
````

## Step 4: Spelling Correction

**Note:** spaCy does not have built-in spelling correction. We'll use `textblob` for this step.

````python
# Install textblob for spelling correction
# Устанавливаем textblob для исправления орфографии
!pip install textblob

# Download required NLTK data for textblob
# Скачиваем необходимые данные NLTK для textblob
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

from textblob import TextBlob
````

````python
# Step 4: Apply spelling corrections on each token
# Шаг 4: Применяем исправление орфографии к каждому токену

# Function to correct spelling of a word
# Функция для исправления орфографии слова
def correct_spelling(word):
    # Skip punctuation marks
    # Пропускаем знаки препинания
    if word in string.punctuation:
        return word
    # Correct spelling using TextBlob
    # Исправляем орфографию с помощью TextBlob
    return str(TextBlob(word).correct())

# Apply correction to each token
# Применяем исправление к каждому токену
corrected_tokens = [correct_spelling(token) for token in tokens]

# Print first 20 corrected tokens
# Выводим первые 20 исправленных токенов
print("First 20 corrected tokens:")
print(corrected_tokens[:20])

# Rebuild the corrected corpus (join tokens with spaces)
# Восстанавливаем исправленный корпус (объединяем токены с пробелами)
corrected_corpus = ' '.join(corrected_tokens)

# Print the corrected corpus (first 500 chars)
# Выводим исправленный корпус (первые 500 символов)
print("\nCorrected corpus (first 500 characters):")
print(corrected_corpus[:500])
````

## Step 5: Part-of-Speech (PoS) Tagging

````python
# Step 5: Apply PoS tags to each corrected token
# Шаг 5: Применяем PoS теги к каждому исправленному токену

# Process the corrected text with spaCy
# Обрабатываем исправленный текст с помощью spaCy
corrected_doc = nlp(corrected_corpus)

# Create list of (token, POS_tag) pairs
# Создаём список пар (токен, POS_тег)
pos_tags = [(token.text, token.pos_, spacy.explain(token.pos_)) for token in corrected_doc]

# Print first 20 tokens with their POS tags
# Выводим первые 20 токенов с их POS тегами
print("Token - POS Tag - Description")
print("-" * 50)
for token, pos_tag, description in pos_tags[:20]:
    print(f"{token:15} -> {pos_tag:5} -> {description}")

# Alternative: Simple POS tags only
# Альтернатива: только POS теги
simple_pos = [(token.text, token.pos_) for token in corrected_doc]
print("\nSimple POS tags (first 20):")
print(simple_pos[:20])
````

## Step 6: Remove Stop Words

````python
# Step 6: Remove stop words from the corrected token list
# Шаг 6: Удаляем стоп-слова из списка исправленных токенов

# Get spaCy's default stop words list
# Получаем стандартный список стоп-слов spaCy
stop_words = spacy.lang.en.stop_words.STOP_WORDS

print(f"Number of stop words in spaCy: {len(stop_words)}")
print(f"Sample stop words: {list(stop_words)[:20]}")

# Remove stop words (keep only words not in stop_words)
# Удаляем стоп-слова (оставляем только слова не из stop_words)
filtered_tokens = [token.text for token in corrected_doc 
                   if token.text.lower() not in stop_words 
                   and token.text not in string.punctuation]

# Print first 20 tokens after removing stop words
# Выводим первые 20 токенов после удаления стоп-слов
print("\nFirst 20 tokens after removing stop words:")
print(filtered_tokens[:20])

# Print number of tokens before and after
# Выводим количество токенов до и после
print(f"\nTokens before removing stop words: {len(corrected_tokens)}")
print(f"Tokens after removing stop words: {len(filtered_tokens)}")
````

## Step 7: Stemming and Lemmatization

````python
# Step 7a: Apply Stemming
# Шаг 7a: Применяем стемминг

# Install nltk for stemming (spaCy doesn't include stemmer)
# Устанавливаем nltk для стемминга (spaCy не включает стеммер)
#!pip install nltk

from nltk.stem import PorterStemmer

# Create stemmer object
# Создаём объект стеммера
stemmer = PorterStemmer()

# Apply stemming to filtered tokens
# Применяем стемминг к отфильтрованным токенам
stemmed_tokens = [stemmer.stem(token) for token in filtered_tokens]

# Print first 20 stemmed tokens
# Выводим первые 20 стеммированных токенов
print("First 20 stemmed tokens:")
print(stemmed_tokens[:20])

# Show comparison: original vs stemmed
# Показываем сравнение: оригинал vs стемминг
print("\nComparison (original -> stemmed):")
for original, stemmed in list(zip(filtered_tokens[:10], stemmed_tokens[:10])):
    print(f"{original:15} -> {stemmed}")
````

````python
# Step 7b: Apply Lemmatization (spaCy has built-in lemmatizer)
# Шаг 7b: Применяем лемматизацию (spaCy имеет встроенный лемматизатор)

# Process filtered tokens as a text
# Обрабатываем отфильтрованные токены как текст
filtered_text = ' '.join(filtered_tokens)
filtered_doc = nlp(filtered_text)

# Extract lemmas from each token
# Извлекаем леммы из каждого токена
lemmatized_tokens = [token.lemma_ for token in filtered_doc]

# Print first 20 lemmatized tokens
# Выводим первые 20 лемматизированных токенов
print("First 20 lemmatized tokens:")
print(lemmatized_tokens[:20])

# Show comparison: original vs lemmatized
# Показываем сравнение: оригинал vs лемматизация
print("\nComparison (original -> lemmatized):")
for original, lemmatized in list(zip(filtered_tokens[:10], lemmatized_tokens[:10])):
    print(f"{original:15} -> {lemmatized}")

# Compare Stemming vs Lemmatization
# Сравниваем стемминг и лемматизацию
print("\n" + "="*50)
print("Stemming vs Lemmatization Comparison:")
print("="*50)
print(f"{'Original':15} {'Stemmed':15} {'Lemmatized':15}")
print("-"*45)
for orig, stem, lem in list(zip(filtered_tokens[:10], stemmed_tokens[:10], lemmatized_tokens[:10])):
    print(f"{orig:15} {stem:15} {lem:15}")
````

## Step 8: Sentence Boundary Detection

````python
# Step 8: Detect sentence boundaries in the given text corpus
# Шаг 8: Определяем границы предложений в текстовом корпусе

# spaCy automatically detects sentence boundaries during processing
# spaCy автоматически определяет границы предложений при обработке

# Process the original text
# Обрабатываем оригинальный текст
sentence_doc = nlp(text_corpus)

# Extract sentences
# Извлекаем предложения
sentences = list(sentence_doc.sents)

# Print total number of sentences
# Выводим общее количество предложений
print(f"Total number of sentences: {len(sentences)}")

# Print each sentence with its number
# Выводим каждое предложение с его номером
print("\n" + "="*50)
print("All sentences:")
print("="*50)
for i, sent in enumerate(sentences, 1):
    print(f"\nSentence {i}:")
    print(f"{sent.text[:200]}")  # First 200 chars of each sentence
    if len(sent.text) > 200:
        print("...")

# Print sentence boundaries (start and end indices)
# Выводим границы предложений (начальные и конечные индексы)
print("\n" + "="*50)
print("Sentence boundaries (character indices):")
print("="*50)
for i, sent in enumerate(sentences, 1):
    print(f"Sentence {i}: starts at {sent.start_char}, ends at {sent.end_char}")
````

````python
# Final Summary - all preprocessing steps results
# Итоговая сводка - результаты всех шагов预处理

print("="*60)
print("TEXT PREPROCESSING SUMMARY")
print("="*60)

print(f"\n1. Original text length: {len(text_corpus)} characters")
print(f"2. Number of tokens (original): {len(tokens)}")
print(f"3. Number of corrected tokens: {len(corrected_tokens)}")
print(f"4. Number of sentences detected: {len(sentences)}")
print(f"5. Number of tokens after stop words removal: {len(filtered_tokens)}")
print(f"6. Number of unique lemmas: {len(set(lemmatized_tokens))}")

print("\n All preprocessing steps completed successfully!")
````
